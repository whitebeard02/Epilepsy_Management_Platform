```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

# Load the data with the new features we created
df_model = pd.read_csv('../data/processed/featured_patient_data.csv', index_col='date', parse_dates=True)

df_model.head()
```

```python
# --- 1. Define the Feature Set (X) ---
# We drop the target variable 'seizure_occurred'
# We also drop 'patient_id' as it's an identifier, not a predictive feature
X = df_model.drop(columns=['seizure_occurred', 'patient_id'])

# --- 2. Define the Target Variable (y) ---
y = df_model['seizure_occurred']

print("Shape of our features (X):", X.shape)
print("Shape of our target (y):", y.shape)
```

```python
# --- Create a chronological train-test split ---
split_percentage = 0.8
split_point = int(len(X) * split_percentage)

X_train, X_test = X[:split_point], X[split_point:]
y_train, y_test = y[:split_point], y[split_point:]

print(f"Training set size: {len(X_train)} days")
print(f"Testing set size: {len(X_test)} days")
print("-" * 20)
print("Training data runs from", X_train.index.min().date(), "to", X_train.index.max().date())
print("Testing data runs from", X_test.index.min().date(), "to", X_test.index.max().date())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

# --- 1. Initialize and Train the Model ---
# We use a random_state for reproducibility
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- 2. Make Predictions on the Test Set ---
predictions = model.predict(X_test)

# --- 3. Evaluate the Model's Performance ---
print("--- Classification Report ---")
# Add the zero_division=0 parameter to handle the warning
print(classification_report(y_test, predictions, target_names=["No Seizure (0)", "Seizure (1)"], zero_division=0))

print("\n--- Confusion Matrix ---")
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test, display_labels=["No Seizure", "Seizure"])
plt.show()
```

```python
import optuna
import xgboost as xgb
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
import numpy as np # <-- Make sure numpy is imported

# --- 1. Define the Objective Function for the NATIVE XGBoost API ---
def objective(trial, data_x, data_y):
    # --- Split data for validation ---
    X_train_split, X_val, y_train_split, y_val = train_test_split(
        data_x, data_y, test_size=0.2, random_state=42, stratify=data_y
    )
    
    # --- Convert data to XGBoost's native DMatrix format ---
    dtrain = xgb.DMatrix(X_train_split, label=y_train_split)
    dval = xgb.DMatrix(X_val, label=y_val)

    # --- Define hyperparameters for the native API ---
    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'verbosity': 0, # Suppress verbose output
        'booster': 'gbtree',
        'lambda': trial.suggest_float('lambda', 1e-8, 1.0, log=True),
        'alpha': trial.suggest_float('alpha', 1e-8, 1.0, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 9),
        'eta': trial.suggest_float('eta', 0.01, 0.3), # 'eta' is the native API's name for learning_rate
    }

    # --- Train the model using the native xgb.train function ---
    model = xgb.train(
        params,
        dtrain,
        evals=[(dval, "validation")],
        early_stopping_rounds=50,
        verbose_eval=False # Suppress verbose output during each round
    )

    # --- Evaluate on the validation set ---
    # Native .predict() returns probabilities; we round them to get 0 or 1
    preds_proba = model.predict(dval)
    preds_binary = np.round(preds_proba)
    f1 = f1_score(y_val, preds_binary, pos_label=1)
    
    return f1

# --- 2. Create and Run the Study ---
study = optuna.create_study(direction='maximize')
study.optimize(lambda trial: objective(trial, X_train, y_train), n_trials=50, show_progress_bar=True)

# --- 3. Print the Best Hyperparameters Found ---
print("\n--- Best Hyperparameters Found ---")
print(study.best_params)
print(f"\nBest F1-score on validation set: {study.best_value:.4f}")
```

```python
# --- 1. Get the best hyperparameters from the study ---
# We need to rename 'eta' to 'learning_rate' for the scikit-learn wrapper if we were to use it,
# but we will stick to the native API for consistency.
best_params = study.best_params

# --- 2. Prepare data for the native API ---
dtrain_full = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# --- 3. Train the final model on the full training set using the native API ---
final_model = xgb.train(
    best_params,
    dtrain_full,
    num_boost_round=1000, # Train for more rounds
    evals=[(dtest, "test")],
    early_stopping_rounds=50,
    verbose_eval=False
)

# --- 4. Evaluate the final model on the unseen test set ---
final_preds_proba = final_model.predict(dtest)
final_preds_binary = np.round(final_preds_proba)

print("--- Final Optimized Model Performance on Test Set ---")
print(classification_report(y_test, final_preds_binary, target_names=["No Seizure (0)", "Seizure (1)"], zero_division=0))

print("\n--- Final Optimized Model Confusion Matrix ---")
# Note: We can't use ConfusionMatrixDisplay.from_estimator with a native model.
# We use from_predictions instead.
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, final_preds_binary)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Seizure", "Seizure"])
disp.plot()
plt.show()
```

```python
import joblib
import shap
import os

print("--- Preparing for Final Model Training ---")

# --- 1. Combine training and testing data ---
X_full = pd.concat([X_train, X_test])
y_full = pd.concat([y_train, y_test])

print(f"Full dataset size for final training: {len(X_full)} days")

# --- 2. Prepare the full dataset in DMatrix format ---
dtrain_full = xgb.DMatrix(X_full, label=y_full)

# --- 3. Train the final model on ALL data using the best hyperparameters ---
print("\n--- Training Final Model on All Data ---")
best_params = study.best_params
final_model = xgb.train(best_params, dtrain_full, num_boost_round=150) # Using a fixed number of rounds
print("Final model training complete.")

# --- 4. Create and save the model artifact ---
model_dir = "../models"
os.makedirs(model_dir, exist_ok=True) # Ensure the 'models' directory exists

model_path = os.path.join(model_dir, "xgb_model_v1.0.json")
final_model.save_model(model_path)
print(f"Final XGBoost model saved to: {model_path}")

# --- 5. Create and save the SHAP explainer artifact ---
print("\n--- Creating and Saving SHAP Explainer ---")

# SHAP requires a background dataset to compute expectations
# Using the full training set is a standard approach
explainer = shap.TreeExplainer(final_model, X_full)

explainer_path = os.path.join(model_dir, "shap_explainer_v1.0.joblib")
joblib.dump(explainer, explainer_path)
print(f"SHAP explainer saved to: {explainer_path}")

print("\n--- Model Finalization Complete ---")
```
