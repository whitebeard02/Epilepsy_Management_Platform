```python
%matplotlib inline
import mne
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import sys

# Add the src directory to the system path to import our custom module
# This allows the notebook to find our 'eeg_features.py' file
sys.path.append('../src')
from eeg_features import calculate_band_power, detect_spikes

# Define constants
DATA_DIR = Path('../data')
RAW_DATA_PATH = DATA_DIR / 'raw'
PROCESSED_DATA_PATH = DATA_DIR / 'processed'

# Ensure the output directory exists
PROCESSED_DATA_PATH.mkdir(parents=True, exist_ok=True)

EEG_FILE = RAW_DATA_PATH / 'chb01_01.edf'
OUTPUT_CSV = PROCESSED_DATA_PATH / 'patient_01_eeg_features.csv'

# Frequency bands (in Hz)
BANDS = {
    'delta': (0.5, 4),
    'theta': (4, 8),
    'alpha': (8, 12),
    'beta': (12, 30),
    'gamma': (30, 100)
}

print(f"Setup complete. Looking for EEG file at: {EEG_FILE}")
```

```python
# We add encoding='latin-1' to fix the TypeError crash during file reading.
raw = mne.io.read_raw_edf(EEG_FILE, preload=True, verbose=False, encoding='latin-1')

# Get sampling frequency
sf = raw.info['sfreq']
print(f"Sampling Frequency: {sf} Hz")

# Select a single EEG channel for this analysis (e.g., 'FP1-F7')
channel = 'FP1-F7'
raw.pick([channel])

# Apply a band-pass filter (e.g., 0.5 Hz to 100 Hz)
raw.filter(l_freq=0.5, h_freq=100, fir_design='firwin', verbose=False)

print(f"Data loaded and pre-processed for channel: {channel}")

# --- START: CORRECTED PLOTTING ---

# We will plot the first 10 seconds of data manually using matplotlib
plot_duration_sec = 10
plot_samples = int(sf * plot_duration_sec)

# Get the data and time vector from the 'raw' object
data, times = raw.get_data(stop=plot_samples, return_times=True)

# Create the plot
plt.figure(figsize=(10, 3))
plt.plot(times, data.T)
plt.xlabel('Time (s)')
plt.ylabel('EEG (V)')
plt.title(f'First 10 Seconds of Channel: {channel}')
plt.grid(True)
plt.show()

# --- END: CORRECTED PLOTTING ---
```

```python
WINDOW_SIZE_SEC = 60 # 1 minute
WINDOW_SAMPLES = int(WINDOW_SIZE_SEC * sf)

all_features = []

# Get the raw data from the selected channel
signal_data = raw.get_data()[0]

# Process the signal in windows
for i in range(0, len(signal_data), WINDOW_SAMPLES):
    window = signal_data[i:i + WINDOW_SAMPLES]

    # Skip last window if it's smaller than the defined size
    if len(window) < WINDOW_SAMPLES:
        continue

    window_features = {'window_start_time': i / sf}

    # 1. Calculate relative power for each band
    for band_name, band_freqs in BANDS.items():
        power = calculate_band_power(window, sf, band_freqs, relative=True)
        window_features[f'rel_power_{band_name}'] = power

    # 2. Detect spikes
    spike_count = detect_spikes(window, sf, prominence=0.7, width=2)
    window_features['spike_count'] = spike_count

    all_features.append(window_features)

features_df = pd.DataFrame(all_features)

print("Feature extraction complete.")
features_df.head()
```

```python
# Define aggregation methods
aggregations = {}
for col in features_df.columns:
    if 'rel_power' in col:
        # For power, we want the mean and standard deviation
        aggregations[col] = ['mean', 'std']
    elif 'spike_count' in col:
        # For spike count, we want the total sum
        aggregations[col] = ['sum']

daily_summary_df = features_df.agg(aggregations)

# Flatten the multi-level column index
daily_summary_df.columns = ['_'.join(col).strip() for col in daily_summary_df.columns.values]
daily_summary_df = daily_summary_df.reset_index(drop=True)

# Add patient and date identifiers (placeholders)
daily_summary_df['patient_id'] = 'chb01'
daily_summary_df['date'] = pd.to_datetime('2025-01-01') # Placeholder date

# Reorder columns for clarity
cols = ['patient_id', 'date'] + [col for col in daily_summary_df.columns if col not in ['patient_id', 'date']]
daily_summary_df = daily_summary_df[cols]

print("Daily aggregation complete.")
daily_summary_df
```

```python
daily_summary_df.to_csv(OUTPUT_CSV, index=False)
print(f"Processed features saved to: {OUTPUT_CSV}")
```
